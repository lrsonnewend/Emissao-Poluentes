```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import csv
```

```python
dados = pd.read_csv('poluentes.csv', sep=';')
dados.head()
```

```python
poluentes = pd.value_counts(dados['Poluente emitido'])
poluentes = poluentes[:4]
poluentes
```

```python
estadoPolu = pd.value_counts(dados['Estado'])
estadoPolu
```

```python
plt.style.use("ggplot")
estadoPolu.plot(kind='bar', subplots=True, label="", title = "Poluição por estados", figsize=(7, 5))
plt.xticks(rotation=90)
```

```python
plt.style.use("ggplot")
poluentes.plot(kind='pie', subplots=True, title="Principais poluentes", figsize=(10, 5))
plt.xticks(rotation=90)
```
