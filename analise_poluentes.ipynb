```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import csv
```

```python
dados = pd.read_csv('poluentes.csv', sep=';')
dados.head()
```

```python
poluentes = pd.value_counts(dados['Poluente emitido'])
poluentes = poluentes[:4]
poluentes
```

```python
estadoPolu = pd.value_counts(dados['Estado'])
estadoPolu
```

```python
plt.style.use("ggplot")
estadoPolu.plot(kind='bar', subplots=True, label="", title = "Poluição por estados", figsize=(7, 5))
plt.xticks(rotation=90)
```

```python
plt.style.use("ggplot")
poluentes.plot(kind='pie', subplots=True, label = "",  fontsize = 13,figsize=(10, 5), autopct='%.2f%%')
plt.title("Porcentagem dos principais poluentes ", fontsize = 16, color = 'darkblue')
```

```python
metodoPolu = pd.value_counts(dados['Detalhe'])
metodoPolu = metodoPolu[:15].sort_values(ascending=[True])
metodoPolu
```

```python
plt.style.use("ggplot")
metodoPolu.plot(kind='barh', subplots=True, label="", figsize=(20, 25), fontsize = 40)
plt.title('15 métodos mais poluentes', fontsize = 45, color = 'darkblue')
plt.xticks(rotation=90)
```

```python
municipioPolu = pd.value_counts(dados['Município'])
municipioPolu = municipioPolu[:30].sort_values(ascending=[True])
municipioPolu
```

```python
plt.style.use("ggplot")
municipioPolu.plot(kind='bar', subplots=True, label="", figsize=(10, 7), fontsize = 14, color = 'darkblue')
plt.title('15 municípios que mais emitem gases poluentes', fontsize = 20)
#plt.xticks(rotation=90)
```
