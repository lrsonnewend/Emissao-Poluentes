```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import csv
```

```python
dados = pd.read_csv('poluentes.csv', sep=';')
dados.head()
```

```python
poluentes = pd.value_counts(dados['Poluente emitido'])
poluentes = poluentes[:4]
poluentes
```

```python
estadoPolu = pd.value_counts(dados['Estado'])
estadoPolu
```

```python
plt.style.use("ggplot")
estadoPolu.plot(kind='bar', subplots=True, label="", title = "Poluição por estados", figsize=(7, 5))
plt.xticks(rotation=90)
```

```python
plt.style.use("ggplot")
poluentes.plot(kind='pie', subplots=True, label = "", title="Principais poluentes", figsize=(10, 5), autopct='%.2f')
plt.xticks(rotation=90)
plt.legend(loc='center left', bbox_to_anchor=(1, 0, 0.5, 1.5))
```

```python
metodoPolu = pd.value_counts(dados['Detalhe'])
metodoPolu = metodoPolu[:15].sort_values(ascending=[True])
```

```python
plt.style.use("ggplot")
metodoPolu.plot(kind='barh', subplots=True, label="", figsize=(9, 15), fontsize = 18)
plt.title('Métodos mais poluentes', fontsize = 20, color = 'darkblue')

plt.xticks(rotation=90)
```
